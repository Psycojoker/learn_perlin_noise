## Noise

First, let's start by implementing the random function given in the article and ploting it.

```python
def random(x):
    x = (x << 13.) ^ x
    return (1.0 - ((x * (x * x * 15731 + 789221) + 1376312589) & 0x7fffffff) / 1073741824.0)
```

```python
plot(map(random, range(40)))
```

Cool, we've got garbage noise which is what we wanted (totally random result).

Notice that all big numbers are supposed to be prime numbers, so to change the random function, changing those prime numbers by other prime numbers is supposed to do the trick.

## Smoothing

Let's implement the given smoothing function and apply it to our noise.

```python
def smooth(x):
    return random(x / 2) + (random(x - 1) / 4.) + (random(x + 1) / 4.)
```

Let's plot it, then plot it in comparison with the random function.

```python
plot(map(smooth, range(40)))
```

## Interpolation

### Linear interpolation

While this is one supposed to be garbish, it's simple, so let's try implementing it to learn, then again, plot it.

```python
def interpolation(x, interpolation_function):
    int_x = int(x)
    fractional_x = x - int_x
    
    a = smooth(int_x)
    b = smooth(int_x + 1)
    
    return interpolation_function(x, a, b)

def linear_interpolation(x, a, b):
    return a * (1 - x) + b * x
```

```python
plot(map(lambda x: interpolation(x, linear_interpolation), range(40)))
```

# Cosine Interpolation

```python
def cosine_interpolation(x, a, b):
    ft = x * 3.1315927
    f = (1 - cos(ft)) * .5
    
    return a * (1 - f) + b * f
```

```python
plot(map(lambda x: interpolation(x, cosine_interpolation), range(40)))
```
